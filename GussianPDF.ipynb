```python
import random 
import matplotlib.pyplot as plt
sig=10
mu=5
pi=3.14

def gaussianPdf(x):
    return(((1/(((2*pi)**0.5)*sig)) * (2.71**(-1*((x-mu)/sig)**2))))
randomX = []
randomY = []
for i in range(0,1000):
    randomX.append(random.randint(0,1000))
    
for i in range(0,len(randomX)):
    randomY.append(gaussianPdf(randomX[i]))
    
plt.plot(randomX,randomY)
plt.show()
```
